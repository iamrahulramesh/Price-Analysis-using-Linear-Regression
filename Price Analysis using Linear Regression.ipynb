```python
df_dr = df.dropna().sort_values("Date")
df_dr
```

```python

all_data = []
```

```python
from datetime import date

df_dr['Date'] = pd.to_datetime(df_dr['Date']) #unless you convert it into date , it keeps throwing error
date_diff = [x - df_dr['Date'].min() for x in df_dr['Date']]
pd.DataFrame(date_diff)
```

```python
df_dr['Date'] =date_diff
```

```python
suburb_dummies = pd.get_dummies(df_dr[['Type','Method']])
```

```python
all_data = df_dr.drop(["Address","Price","Date", "SellerG","Suburb","Type","Method","CouncilArea","Regionname"],
                      axis =1).join(suburb_dummies)
```

```python
all_data
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
lr = LinearRegression()
```

```python
X = all_data
y =df_dr['Price']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
```

```python
lr.fit(X_train,y_train)
```

```python
print(lr.intercept_)
```

```python
X.columns
```

```python
coef_df = pd.DataFrame(lr.coef_,X.columns,columns =['Coefficient'])
ranked_suburbs = coef_df.sort_values('Coefficient',ascending = False)
```

```python
coef_df
```

```python
ranked_suburbs
```

```python
y_pred = lr.predict(X_test)
y_pred
```

```python

plt.scatter(y_test,y_pred)
plt.ylim([200000,1000000])
plt.xlim([200000,1000000])
```

```python
fig,ax = plt.subplots(figsize = (15,15))
sns.distplot(y_test-y_pred,bins = 50)
```

```python
from sklearn import metrics
```

```python
print('Mean Absolute Error : ',round(metrics.mean_absolute_error(y_test,y_pred)))
print('Mean Squared Error: ',round(metrics.mean_squared_error(y_test,y_pred)))
print("Root Mean Squared Error: ",round(np.sqrt(metrics.mean_squared_error(y_test,y_pred))))
```
